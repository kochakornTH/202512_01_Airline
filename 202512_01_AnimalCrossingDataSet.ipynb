```python
import pandas as pd

import kagglehub
import os
```

```python
path = kagglehub.dataset_download("jessicali9530/animal-crossing-new-horizons-nookplaza-dataset")

fileList = os.listdir(path)
print(fileList)
```

```python
fileDict = {}
for i in fileList:
  df = pd.read_csv(os.path.join(path,i))
  fileDict[i]= df.shape
```

```python
df_Data = pd.DataFrame(fileDict).T
```
